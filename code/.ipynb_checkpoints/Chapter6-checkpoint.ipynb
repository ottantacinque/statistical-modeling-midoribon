## 6.3 二項分布で表現する「あり・なし」カウントデータ

```python
# 上限N=8で、正規確率qが{0.1, 0.3, 0.8}である二項分布の確率分布を描画する
q_list = [0.1,0.3,0.8]
yi_list = np.arange(0, 9)

p_dicts =dict()
for q in q_list:
    prob_list = binom.pmf(yi_list, 8, q)
    plt.plot(yi_list, prob_list, marker="o", label=f'q={q}')
    
plt.xlabel('$y_i$', fontsize=15)
plt.ylabel('確率 $p(y_i|8,q)$', fontsize=15)
plt.legend(loc='best')
plt.show()
```

## 6.5 交互作用項の入った線形予測子

```python
# 交互作用を入れた統計モデルのパラメーター推定
model = smf.glm('y + I(N-y)~ x*f',data=d,family=sm.families.Binomial())
result = model.fit()
print(result.summary())
print('AIC: ', round(result.aic,3))

"""R
glm(cbind(y, N - y) ~ x * f, family = binomial, data = d)
"""
```

## 6.6 割り算値の統計モデリングはやめよう

```python
# オフセット項を指定したモデルでのパラメーター推定
df = pd.read_csv("data4b.csv")
model = smf.glm('y~x',offset=np.log(df["A"]), data=df, family=sm.families.Poisson())
result = model.fit()
print(result.summary())

"""R
glm(y ~ x, offset = log(A), family = poisson, data = d)
d <- read.csv("data4b.csv")
"""
```

```python
# データを図示してみる
df_predict = pd.DataFrame()
df_predict["A"] = np.arange(0.1,20,0.1)
x_list = [0.1,0.3,0.5,0.7,0.9]

for x in x_list:
    df_predict["x"] = x
    predict_y = result.predict(df_predict, offset=np.log(df_predict["A"]))
    plt.plot(df_predict['A'], predict_y, alpha=0.8, label=f'x={x}')
    
plt.scatter(df['A'],df['y'], c=df['x'])
    
plt.xlim(0,18)
plt.ylim(0,100)
plt.xlabel("調査地の面積 $A_i$", fontsize=15)
plt.ylabel("植物の個体数", fontsize=15)

plt.colorbar()
plt.legend()
```

## 正規分布とその尤度

```python
# 正規位分布の密度関数を図示してみる
y = np.arange(-5,5,0.1)
plt.plot(y, sct.norm.pdf(y, loc=0, scale=1), label=r'$\mu=0, \sigma=1$')
plt.plot(y, sct.norm.pdf(y, loc=0, scale=3), label=r'$\mu=0, \sigma=3$')
plt.plot(y, sct.norm.pdf(y, loc=2, scale=1), label=r'$\mu=2, \sigma=1$')

plt.legend()
plt.show()


"""R
y <- seq(-5, 5, 0.1)
plot(y, dnorm(y, mean=0, sd=1), type="l")
"""
```

```python
# 1.2<=y<=1.8 の範囲の確率を計算する
print(sct.norm.cdf(x=1.8,loc=0,scale=1) - sct.norm.cdf(x=1.2,loc=0,scale=1))

"""R
pnorm(1.8, 0, 1) - pnorm(1.2, 0, 1)
"""

# 長方形の面積として上記の確率を近似計算する
print(sct.norm.pdf(x=((1.2+1.8)/2),loc=0,scale=1) * (1.8 -1.2))

"""R
dnorm(1.5, 0, 1) * 0.6
"""
```

## 6.8 ガンマ分布のGLM

```python
# データの読み込みと図示
df = pd.read_csv("./d.csv",index_col=0)
plt.scatter(df['x'],df['y'])
plt.xlabel("葉重量 $x_i$", fontsize=15)
plt.ylabel("花重量 $y_i$", fontsize=15)
plt.show()

"""R
d <- read.csv("d.csv")
plot(d$x, d$y, type="p")
"""
```

```python
# ガンマ分布を用いてパラメータ推定する
model = smf.glm('y~np.log(x)', data=df, family=sm.families.Gamma(link=sm.families.links.log()))
result = model.fit()
print(result.summary())

"""R
glm(y ~ log(x), family=Gamma(link="log"), data=d)
"""
```

```python
#結果を図示する
df_predict = pd.DataFrame()
df_predict["x"] = np.arange(0.01,0.9,0.01)

df_predict["pred_y"] = result.predict(df_predict["x"])
plt.scatter(df["x"],df["y"])
plt.plot(df_predict["x"],df_predict["pred_y"],c="red")

plt.xlabel("葉重量 $x_i$", fontsize=15)
plt.ylabel("花重量 $y_i$", fontsize=15)
plt.show()
```
